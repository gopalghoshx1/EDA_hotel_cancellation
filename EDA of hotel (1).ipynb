# Data Analytics and Visualization

```python
cancel_perc=df['is_canceled'].value_counts(normalize=True)
cancel_perc
```

```python
plt.figure(figsize=(3,5))
plt.title('reservation status count')
plt.bar(('Not canceled','canceled'),df['is_canceled'].value_counts(),width=0.7,edgecolor='k')
plt.show()
```

```python
plt.figure(figsize= (8,5))
ax1=sns.countplot(x='hotel',hue='is_canceled', data = df ,palette ='Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(['Not canceled','canceled'],bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('Hotel', size=12)
plt.ylabel('Number of reservation',size=12)
plt.show()
```

```python
resort_Hotel=df[df['hotel']=='Resort Hotel']
resort_Hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_Hotel = df[df['hotel']=='City Hotel']
city_Hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_Hotel=(city_Hotel.groupby('reservation_status_date')[['adr']].mean())
resort_Hotel=(resort_Hotel.groupby('reservation_status_date')[['adr']].mean())
```

```python
plt.figure(figsize=(15,6))
plt.plot( city_Hotel.index,city_Hotel['adr'], label='City_Hotel')
plt.plot(resort_Hotel.index,resort_Hotel['adr'], label='Resort_Hotel')
plt.title('Average Daily Rate in City and Resort Hotel',fontsize= 20)
plt.xlabel('Date', fontsize= 15)
plt.ylabel('Rate',fontsize= 15)
plt.legend(loc="upper left",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='reservation_Month',hue='is_canceled', data = df ,palette ='Reds')
plt.title('Reservation status of every month', fontsize=18)
plt.legend(['Not canceled','canceled'], fontsize= 12)
plt.show()
```

```python
canceled_record_df= df[df['is_canceled']==1].groupby('reservation_Month')[['adr']].sum().reset_index()
canceled_record_df
```

```python
plt.figure(figsize=(12,5))
sns.barplot('reservation_Month','adr', data = canceled_record_df)
plt.title('Canceled Reservation Rate status of every month', fontsize=18)
plt.show()
```

```python
canceled= df[df['is_canceled']==1]
plt.figure(figsize=(14,8))
plt.title('Top 10 countrys of Cancelation', fontsize=15)
plt.pie(canceled['country'].value_counts()[:10], labels= canceled['country'].value_counts()[:10].index, autopct='%.2f')
plt.show()
```

```python
nt_canceled = df[df['is_canceled']==0]
nt_cancel_df=nt_canceled[(nt_canceled['reservation_status_date'] > '2016') & (nt_canceled['reservation_status_date'] < '2017')]
cleared_nt_cancel_df =( nt_cancel_df.groupby('reservation_status_date')[['adr']].mean())


cancel_df=canceled[(canceled['reservation_status_date'] > '2016') & (canceled['reservation_status_date'] < '2017')]
cleared_cancel_df =(cancel_df.groupby('reservation_status_date')[['adr']].mean())
```
